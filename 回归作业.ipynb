```python
import pandas as pd
import numpy as np
from sklearn import datasets
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.model_selection import train_test_split
```

```python
x_df=pd.DataFrame(datasets.load_boston()['data'],columns=datasets.load_boston()['feature_names']) #X
y_df=pd.DataFrame(datasets.load_boston()['target'],columns=['y']) #Y
df=x_df.join(y_df)
```

```python
plt.rcParams['font.sans-serif']=['Simhei']
plt.rcParams['axes.unicode_minus']=False
```

```python
# RM
plt.scatter(x_df['RM'], y_df)
plt.title('住宅房间数与房价散点图')
plt.xlabel('住宅房间数')
plt.ylabel('房价')
plt.show()
```

```python
#DIS
plt.scatter(x_df['DIS'], y_df)
plt.title('距离5个就业中心的平均距离与房价散点图与房价散点图')
plt.xlabel('距离5个就业中心的平均距离')
plt.ylabel('房价')
plt.show()
```

```python
#PTRATIO
plt.scatter(x_df['PTRATIO'], y_df)
plt.title('学生教师比例与房价散点图')
plt.xlabel('学生教师比例')
plt.ylabel('房价')
plt.show()
```

```python
#LSTAT
plt.scatter(x_df['LSTAT'], y_df)
plt.title('低收入阶层占比与房价散点图')
plt.xlabel('低收入阶层占比')
plt.ylabel('房价')
plt.show()
```

**划分训练集测试集**

```python
feature_train, feature_test, target_train, target_test = train_test_split(x_df, y_df, test_size=0.1, random_state=42)
```

```python
feature_train = feature_train[['RM','DIS','PTRATIO','LSTAT']]
```

```python
feature_train
```

```python
target_train
```

```python
feature_test = feature_test[['RM','DIS','PTRATIO','LSTAT']]
feature_test
```

```python
target_test
```

**定义模型训练**

```python
lr = linear_model.LinearRegression()
lr.fit(feature_train,target_train)
```

**输出模型参数**

```python
print('普通线性回归斜率:{}'.format(lr.coef_))
print('普通线性回归偏置:{}'.format(lr.intercept_))
```

**模型预测**

```python
y_pred = lr.predict(feature_test)
```

```python
y_pred
```

```python
plt.xlim([0,50])
plt.plot( range(len(target_test)), target_test, 'r', label='y_target')
plt.plot( range(len(y_pred)), y_pred, 'g--', label='y_predict' )
plt.title('sklearn: Linear Regression')
plt.legend()
plt.show()
```

**预测真实数据**

```python
feature = np.array([[8,2,12,22]])
y_pred = lr.predict(feature)
print('预测房价为: %.1f'%y_pred[0][0])
```
