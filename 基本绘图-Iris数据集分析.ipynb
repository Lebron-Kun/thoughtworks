```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

**1、读取数据**

```python
iris_data=pd.read_csv("iris.csv")
```

**2、查看数据**

```python
iris_data.head()   #Id: 鸢尾花编号 ；SepaLengthCm: 花萼长度, 单位cm ；SepalWidthCm: 花萼宽度, 单位cm ；PetalLengthCm: 花瓣长度, 单位cm ；PetalWidthCm; 花瓣宽度, 单位cm ；Species: 鸢尾花种类.
```

**3、设置绘图参数**

```python
fig,ax=plt.subplots(1,2,figsize=(15, 8))

# 绘制 花瓣特征与鸢尾花类型分布关系图
iris_data[iris_data['Species']=='Iris-setosa'].plot(kind='scatter',x='PetalLengthCm',y='PetalWidthCm',ax=ax[0],color='steelblue',title='花瓣特征与鸢尾花类型分布', label='Setosa')
iris_data[iris_data['Species']=='Iris-versicolor'].plot(kind='scatter',x='PetalLengthCm',y='PetalWidthCm',ax=ax[0],color='darkorange',marker='^',label='Versicolor')
iris_data[iris_data['Species']=='Iris-virginica'].plot(kind='scatter',x='PetalLengthCm',y='PetalWidthCm',ax=ax[0],color='green',marker='s',label='Virginica')
# 绘制 花萼特征与鸢尾花类型分布的关系图
iris_data[iris_data['Species']=='Iris-setosa'].plot(kind='scatter',x='SepalLengthCm',y='SepalWidthCm',ax=ax[1],color='steelblue', title='花萼特征与鸢尾花类型分布', label='Setosa')
iris_data[iris_data['Species']=='Iris-versicolor'].plot(kind='scatter',x='SepalLengthCm',y='SepalWidthCm',ax=ax[1],color='darkorange',marker='^', label='Versicolor')
iris_data[iris_data['Species']=='Iris-virginica'].plot(kind='scatter',x='SepalLengthCm',y='SepalWidthCm',ax=ax[1],color='green',marker='s',label='Virginica')

plt.show()
plt.close()
```
